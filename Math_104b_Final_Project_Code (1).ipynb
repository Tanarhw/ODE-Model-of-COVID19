```python
#RUNGE KUTTA ORDER 4 CODE

import scipy as sp

import numpy as np

import matplotlib.pylab as plt



N=2489802 #total population of city

I=10. #infected individuals at time t

S=(N-I) #susceptible individuals at time t

R=0. #recovered individuals at time t

r=1./10 #transmission rate of virus

a=1./10 #rate of recovery, 1/10 -> 10 days to recover

A=N #inital value (not sure what this means)

u=0.00053#death rate 3.4%

h=1/2



n_steps = 100



S = np.zeros(n_steps+1)

I = np.zeros(n_steps+1)

R = np.zeros(n_steps+1)

print(S.shape)

S[0] = (N-I[0])

I[0] = 10

R[0] = 0



for i in range(n_steps):



    k1 = A - u*S[i] - (r/N)*I[i]*S[i]

    l1 = (r*I[i]/N) * S[i] - (u+a) * I[i]

    m1 = a * I[i] - u * R[i]



    k2 = A - u*(S[i] + k1*h/2) - (r/N) * (I[i] + l1*h/2) * (S[i] + k1*h/2)

    l2 = (r/N) * (I[i] + l1*h/2) * (S[i] + k1*h/2) - (u+a)*(I[i] + l1*h/2)

    m2 = a * (I[i] + l1*h/2) - u*(R[i] + m1*h/2)



    k3 = A - u*(S[i] + k2*h/2) - (r/N) * (I[i] + l2*h/2) * (S[i] + k2*h/2)

    l3 = (r/N) * (I[i] + l2*h/2) * (S[i] + k2*h/2) - (u+a) * (I[i] + l2*h/2)

    m3 = a*(I[i] + l2*h/2) - u*(R[i] + m2*h/2)



    k4 = A - u*(S[i] + k3*h) - (r/N)*(I[i] + l3*h) * (S[i] + k3*h)

    l4 = (r/N) * (I[i] + l3*h) * (S[i] + k3*h) - (u+a) * (I[i] + l3*h)

    m4 = a * (I[i] + l3*h) - u * (R[i] + m3*h)



    S[i+1] = S[i] + (1.0/6) * (k1 + 2*k2 + 2*k3 + k4) * h

    I[i+1] = I[i] + (1.0/6) * (l1 + 2*l2 + 2*l3 + l4) * h

    R[i+1] = R[i] + (1.0/6) * (m1 + 2*m2 + 2*m3 + m4) * h








plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

plt.xlabel('Time in days')
plt.ylabel('Total Population (in millions)')
plt.title('COVID-19 South Korea RK-4')
S_line, = plt.plot(S, label='S')
I_line, = plt.plot(I, label='I')
R_line, = plt.plot(R, label='R')
plt.legend(handles=[S_line, I_line, R_line])
plt.show()

#print(S)
np.argmax(I)
max(I)

#EULERS METHOD CODE

N=2489802 #total population of city

r=1./10 #transmission rate of virus

a=1./10 #rate of recovery, 1/10 -> 10 days to recove

A=N #inital value (not sure what this means)

u=0.00053 #death rate 3.4%

h=1/2


n_steps = 100

S = np.zeros(n_steps+1)

I = np.zeros(n_steps+1)

R = np.zeros(n_steps+1)

#print(S.shape)

S[0] = (N-I[0])
I[0] = 10

R[0] = 0



for i in range (n_steps):
   
    k1 = A - u*S[i] - (r/N)*I[i]*S[i]

    l1 = (r*I[i]/N) * S[i] - (u+a) * I[i]

    m1 = a * I[i] - u * R[i]
    
    S[i+1] = S[i] + k1 * h
 
    I[i+1] = I[i] + l1 * h

    R[i+1] = R[i] + m1 * h

 
    

plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

plt.xlabel('Time in days')
plt.ylabel('Total Population (in millions)')
plt.title('COVID-19 South Korea Eulers')
S_line, = plt.plot(S, label='S')
I_line, = plt.plot(I, label='I')
R_line, = plt.plot(R, label='R')
plt.legend(handles=[S_line, I_line, R_line])
plt.show()
np.argmax(I)
max(I)
```
